```python
import ast
import pprint

# 예시: original_dict 변수에 제공해주신 데이터가 있다고 가정합니다.
original_dict = store_logs

def parse_trade_log(log_str):
    """
    로그 문자열이 단순 문자열('관망' 등)인지, 아니면 내부에 주문내역이 담긴 딕셔너리 문자열인지를 구분하여,
    주문 내역이 있으면 "매수/매도 | 수량 | 가격" 문자열을 반환합니다.
    """
    # 양쪽에 따옴표가 있는 경우 제거
    log_str = log_str.strip().strip('"').strip("'")
    if log_str == "관망":
        return None  # 주문 없음

    # 내부에 딕셔너리 형식의 문자열이 있는지 확인
    try:
        # ast.literal_eval를 사용하여 문자열을 딕셔너리로 변환 시도
        parsed = ast.literal_eval(log_str)
        if isinstance(parsed, dict) and "log" in parsed:
            return parsed["log"].strip()
    except Exception:
        pass
    # 그 외의 경우 단순 문자열로 판단
    return log_str

# 결과를 저장할 딕셔너리 (날짜별 → 티커별 → 주문내역 리스트)
daily_trades = {}

# original_dict의 각 키는 "YYYYMMDD_t_2_t_4_trader_log" 형태입니다.
for day_key, day_log_str in original_dict.items():
    # 날짜 추출 (예: '20230102')
    date = day_key.split('_')[0]
    if date not in daily_trades:
        daily_trades[date] = {}

    try:
        # 내부 문자열을 딕셔너리로 변환합니다.
        day_log = ast.literal_eval(day_log_str)
    except Exception as e:
        print(f"{day_key} 파싱 에러: {e}")
        continue

    # day_log는 {시간대: {티커: 로그, ...}, ...} 형태
    for time_interval, ticker_dict in day_log.items():
        for ticker, log_str in ticker_dict.items():
            trade = parse_trade_log(log_str)
            if trade:  # 주문 내역이 있는 경우
                # 주문 내역은 "매수 | 수량 | 가격" 또는 "매도 | 수량 | 가격" 형식으로 되어 있다고 가정
                parts = [p.strip() for p in trade.split("|")]
                if len(parts) == 3:
                    action, qty, price = parts
                else:
                    action, qty, price = trade, None, None

                # 날짜별, 티커별로 주문내역을 저장
                if ticker not in daily_trades[date]:
                    daily_trades[date][ticker] = []
                daily_trades[date][ticker].append({
                    "time": time_interval,
                    "action": action,
                    "qty": qty,
                    "price": price
                })

# 시간 순으로 정렬(옵션)
for date in daily_trades:
    for ticker in daily_trades[date]:
        daily_trades[date][ticker].sort(key=lambda x: x["time"])

daily_trades
```

```python
# ------------------------------------------------------------------------------
# [3] 시뮬레이션 함수 정의
# ------------------------------------------------------------------------------
def to_number(num_str):
    """
    숫자 문자열(예:"25,700")을 int 또는 float로 변환.
    소수점이 있으면 float, 없으면 int로 변환합니다.
    """
    try:
        # 쉼표 제거 후 소수점 여부로 변환
        num_str = num_str.replace(",", "")
        if '.' in num_str:
            return float(num_str)
        else:
            return int(num_str)
    except Exception:
        return 0

def simulate_quarter(orders, starting_price):
    """
    한 종목에 대해, 주어진 주문(orders) 리스트를 시간순으로 적용하여
    포트폴리오 시뮬레이션을 진행합니다.
    
    가정:
      - 분기 시작 시 해당 종목 1주를 시작 가격에 보유.
      - 초기 현금은 0.
      - 각 주문은 { "time": 시간, "action": "매수"/"매도", "qty": 수량, "price": 가격 } 형태.
    
    최종 포트폴리오 가치는: (남은 주식수 * 마지막 거래가격) + 현금
    수익률 = (최종 포트폴리오 가치 - 시작가) / 시작가
    (여기서 시작가는 1주 보유 기준이므로, 시작 가치 = starting_price)
    """
    position = 1       # 초기 1주 보유
    cash = 0.0         # 초기 현금
    last_price = starting_price  # 마지막 거래가격 (없으면 시작가)
    
    for order in orders:
        # 문자열 수량, 가격을 숫자로 변환
        qty = to_number(order["qty"]) if order.get("qty") else 0
        price = to_number(order["price"]) if order.get("price") else 0
        if price: 
            last_price = price  # 주문 시점의 가격을 최신 가격으로 갱신
        
        if order["action"] == "매수":
            position += qty
            cash -= qty * price
        elif order["action"] == "매도":
            position -= qty
            cash += qty * price

    final_value = cash + position * last_price
    return_rate = (final_value - starting_price) / starting_price
    return return_rate, final_value, position, cash, last_price

def simulate_quarter_for_ticker(trades_by_date, ticker, starting_price):
    """
    해당 분기(여러 날짜의 거래내역이 trades_by_date에 저장됨)에서
    특정 ticker에 대한 모든 주문을 날짜 순(그리고 각 날짜 내 시간순)으로 모아서
    시뮬레이션을 진행합니다.
    """
    orders = []
    for date in sorted(trades_by_date.keys()):
        day_trades = trades_by_date[date]
        if ticker in day_trades:
            # 각 날짜에서 해당 ticker의 주문 내역은 이미 시간순 정렬되었다고 가정
            orders.extend(day_trades[ticker])
    return simulate_quarter(orders, starting_price)

# ------------------------------------------------------------------------------
# [4] 분기별 시뮬레이션 실행
# ------------------------------------------------------------------------------
# quarter_results: 분기별로 각 ticker의 시뮬레이션 결과를 저장하는 딕셔너리
quarter_results = {}

for quarter, trades_by_date in quarters.items():
    quarter_results[quarter] = {}
    # 해당 분기의 시작가격 정보 (미리 quarter_starting_prices에서 제공)
    starting_prices = quarter_starting_prices.get(quarter, {})
    for ticker, start_price in starting_prices.items():
        # 각 ticker에 대해 분기의 주문들을 모아 시뮬레이션 진행
        result = simulate_quarter_for_ticker(trades_by_date, ticker, start_price)
        return_rate, final_value, position, cash, last_trade_price = result
        quarter_results[quarter][ticker] = {
            "return_rate": return_rate,      # 예: 0.05 = 5% 수익률
            "final_value": final_value,       # 최종 포트폴리오 가치 (현금+주식가치)
            "final_position": position,       # 최종 주식수
            "final_cash": cash,
            "last_trade_price": last_trade_price
        }

# 결과 출력
pprint.pprint(quarter_results)
```

```python
initial_capital = 1_000_000  # 100만원

# portfolio_quarter_results에 분기별, 티커별 투자 결과(할당금액, 최종 가치, 수익 등)를 저장합니다.
portfolio_quarter_results = {}

for quarter, ticker_results in quarter_results.items():
    portfolio_quarter_results[quarter] = {}
    
    # 해당 분기에 투자할 티커 리스트 (quarter_starting_prices에 기록된 티커 사용)
    tickers = list(quarter_starting_prices.get(quarter, {}).keys())
    n = len(tickers)
    if n == 0:
        continue  # 해당 분기에 투자할 종목이 없으면 건너뜁니다.
    
    # 각 종목에 균등 배분 (예: 100만원 ÷ 종목수)
    allocation_per_ticker = initial_capital / n

    total_final_value = 0
    total_profit = 0

    for ticker in tickers:
        starting_price = quarter_starting_prices[quarter].get(ticker)
        # starting_price가 0인 경우(예: 거래가 없는 종목)는 건너뜁니다.
        if not starting_price or starting_price == 0:
            continue

        sim_result = ticker_results.get(ticker)
        if sim_result is None:
            # 해당 티커에 거래가 없으면, 투자 후에도 변동이 없다고 가정
            ticker_final_value = allocation_per_ticker
            ticker_profit = 0
        else:
            # 시뮬레이션 결과는 1주 기준이므로, 투자금액에 따라 구매 가능한 주식 수를 계산합니다.
            # 구매 가능한 주식 수 = 할당금액 / 시작 가격
            shares = allocation_per_ticker / starting_price

            # 1주 기준 최종 가치(sim_result["final_value"])에 구매 주식수를 곱해 최종 가치를 계산합니다.
            ticker_final_value = shares * sim_result["final_value"]
            ticker_profit = ticker_final_value - allocation_per_ticker

        ticker_return_rate = ticker_profit / allocation_per_ticker

        portfolio_quarter_results[quarter][ticker] = {
            "allocation": allocation_per_ticker,
            "starting_price": starting_price,
            "final_value": ticker_final_value,
            "profit": ticker_profit,
            "profit_rate": ticker_return_rate
        }
        total_final_value += ticker_final_value
        total_profit += ticker_profit

    portfolio_quarter_results[quarter]["_total"] = {
        "initial_capital": initial_capital,
        "total_final_value": total_final_value,
        "total_profit": total_profit,
        "total_return_rate": total_profit / initial_capital
    }

# 결과 출력 (pprint를 사용하여 보기 좋게 출력)
pprint.pprint(portfolio_quarter_results)
```

```python
import json
import pandas as pd
# stock 모듈의 get_market_ohlcv 함수가 존재한다고 가정합니다.
from pykrx import stock

def to_number(num_str):
    """
    문자열로 표현된 숫자(예:"25,700")를 쉼표 제거 후 숫자(int 또는 float)로 변환합니다.
    """
    try:
        num_str = num_str.replace(",", "")
        if '.' in num_str:
            return float(num_str)
        else:
            return int(num_str)
    except Exception:
        return 0

def simulate_day(orders, starting_price):
    """
    하루 동안의 주문(orders)을 순차적으로 반영하여,
    초기 1주 보유 상태에서 최종 포트폴리오 가치와 수익률(%)을 계산합니다.
    
    가정:
      - 초기 포지션: 1주 보유, 초기 현금 0.
      - 각 주문은 {"time": 시간, "action": "매수"/"매도", "qty": 수량, "price": 가격} 형태.
      - 마지막 거래 가격(last_trade_price)이 최종 주가로 사용됨.
      
    반환값:
      profit_rate: (최종 가치 - 시작가) / 시작가 * 100 (수익률 %)
      final_value: 최종 포트폴리오 가치 (현금 + 남은 주식 가치)
    """
    position = 1      # 초기 1주 보유
    cash = 0.0        # 초기 현금
    last_price = starting_price  # 거래가 없으면 시작 가격이 최종 가격으로 사용됨
    
    for order in orders:
        qty = to_number(order.get("qty", "0"))
        price = to_number(order.get("price", "0"))
        if price:
            last_price = price  # 거래 시 갱신
        if order["action"] == "매수":
            position += qty
            cash -= qty * price
        elif order["action"] == "매도":
            position -= qty
            cash += qty * price

    final_value = cash + position * last_price
    profit_rate = (final_value - starting_price) / starting_price * 100
    return profit_rate, final_value

def get_daily_starting_prices(ticker, start_date, end_date):
    """
    지정한 티커와 기간에 대해 OHLCV 데이터를 조회하여,
    각 날짜의 시가를 'YYYYMMDD' 형식의 문자열 키로 반환합니다.
    
    예시 반환:
    {
        "20230102": 25000,
        "20230103": 26000,
        ...
    }
    """
    df = stock.get_market_ohlcv(start_date, end_date, ticker)
    daily_prices = {}
    for date, row in df.iterrows():
        # DataFrame의 인덱스가 pd.Timestamp인 경우 "YYYYMMDD" 형식의 문자열로 변환
        if isinstance(date, pd.Timestamp):
            date_str = date.strftime("%Y%m%d")
        else:
            date_str = str(date)
        daily_prices[date_str] = row["시가"]
    return daily_prices

# --------------------------------------------------------------------------
# [자동화] daily_starting_prices를 외부 API 호출로 자동 생성하기
# --------------------------------------------------------------------------
# daily_trades의 날짜 범위를 기준으로 전체 시작일과 종료일을 결정합니다.
all_days = sorted(daily_trades.keys())
overall_start_date = min(all_days)  # 예: "20230102"
overall_end_date = max(all_days)    # 예: "20230103"

# daily_trades에 포함된 모든 티커 목록 추출
all_tickers = set()
for day, tickers in daily_trades.items():
    all_tickers.update(tickers.keys())

# 각 티커별로 OHLCV 정보를 조회해 일자별 시가를 자동으로 생성합니다.
computed_starting_prices = {}
for ticker in all_tickers:
    computed_starting_prices[ticker] = get_daily_starting_prices(ticker, overall_start_date, overall_end_date)
    
# --------------------------------------------------------------------------
# 일별 수익률(log) 계산 (computed_starting_prices 사용)
# --------------------------------------------------------------------------
daily_profit_rate_log = {}

for day, ticker_orders in daily_trades.items():
    daily_profit_rate_log[day] = {}
    profit_rates = []
    for ticker, orders in ticker_orders.items():
        # computed_starting_prices[ticker]는 { "YYYYMMDD": 시가, ... } 형태입니다.
        if ticker in computed_starting_prices and day in computed_starting_prices[ticker]:
            starting_price = computed_starting_prices[ticker][day]
        else:
            # 시작 가격이 없으면 첫 주문 가격을 사용 (혹은 기본값 1)
            if orders and "price" in orders[0]:
                starting_price = to_number(orders[0]["price"])
            else:
                starting_price = 1
        profit_rate, final_value = simulate_day(orders, starting_price)
        daily_profit_rate_log[day][ticker] = round(profit_rate, 2)
        profit_rates.append(profit_rate)
    # 각 날짜의 _total은 티커별 수익률의 평균값 (원하는 방식에 맞게 조정 가능)
    if profit_rates:
        daily_profit_rate_log[day]["_total"] = round(sum(profit_rates) / len(profit_rates), 2)
    else:
        daily_profit_rate_log[day]["_total"] = 0.0

# --------------------------------------------------------------------------
# 결과 JSON 파일로 저장
# --------------------------------------------------------------------------
with open("daily_profit_rate_log.json", "w", encoding="utf-8") as f:
    json.dump(daily_profit_rate_log, f, ensure_ascii=False, indent=4)

print("daily_profit_rate_log.json 파일이 저장되었습니다.")
```
