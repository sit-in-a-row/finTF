```python
from datetime import datetime, timedelta

# 시작 날짜와 종료 날짜 설정
start_date = datetime(2019, 1, 1)
end_date = datetime(2019, 12, 31)

# ticker 심볼 설정
ticker = "MSFT"

# URL 리스트 초기화
urls = []

# 날짜 범위 내에서 매일 URL 생성
current_date = start_date
while current_date <= end_date:
    # 다음 날짜
    next_date = current_date + timedelta(weeks=1)
    
    # URL 생성
    url = f'https://seekingalpha.com/symbol/{ticker}/news?from={current_date.strftime("%Y-%m-%d")}T15%3A00%3A00.000Z&to={next_date.strftime("%Y-%m-%d")}T14%3A59%3A59.999Z'
    
    # URL 리스트에 추가
    urls.append(url)
    
    # 다음 날짜로 이동
    current_date = next_date

# 결과 출력
for url in urls:
    print(url)
```
